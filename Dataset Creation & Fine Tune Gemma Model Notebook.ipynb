```python
import os
import transformers
import torch
from google.colab import userdata
from trl import SFTTrainer
from peft import LoraConfig
from transformers import AutoTokenizer, AutoModelForCausalLM
from transformers import BitsAndBytesConfig, GemmaTokenizer
```

```python
# HuggingFace token
os.environ["HF_TOKEN"] = userdata.get('HF_TOKEN')
```

```python
# Original Gemma model & using 4-bit configuration
model_id = "google/gemma-2b"
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

```python
# Load tokenizer and model using current configuration
tokenizer = AutoTokenizer.from_pretrained(model_id, token=os.environ['HF_TOKEN'])
model = AutoModelForCausalLM.from_pretrained(model_id,
                                             quantization_config=bnb_config,
                                             device_map={"":0},
                                             token=os.environ['HF_TOKEN'])
```

```python
os.environ["WANDB_DISABLED"] = "false"
```

```python
# LoRA confuration with dimension = 8
lora_config = LoraConfig(
    r = 8,
    target_modules = ["q_proj", "o_proj", "k_proj", "v_proj",
                      "gate_proj", "up_proj", "down_proj"],
    task_type = "CAUSAL_LM",
)
```

```python
# "Subject": "Application crashing",
# "Description": "An application keeps crashing. I've tried restarting my computer and reinstalling the application, but it's still crashing. I'm using a Microsoft Office application on a Windows 11 computer.",
# "Status": "Open",
# "Priority": "Medium",
```

```python
# Text/data formation for model
def formatting_func(example):
    text = f"Description: {example['Description'][0]}\nPriority: {example['Priority'][0]}\Subject: {example['Subject'][0]}"
    return [text]
```

```python
!pip install -q evaluate
```

```python
# Evaliate  Method
import evaluate
import numpy as np

metric = evaluate.load("accuracy")

def compute_metrices(eval_pred):
  logits, lebels = eval_pred
  predictions = np.argmax(logits, axis = -1)
  return metric.compute(predictions = predictions, references = lebels)
```

```python
# Additional Configuration
tokenizer.padding_side = 'right'
```

```python
# Using SFT trainer
trainer = SFTTrainer(
    model=model,
    train_dataset=dataset,
    args=transformers.TrainingArguments(
        per_device_train_batch_size=1,
        gradient_accumulation_steps=4,
        warmup_steps=2,
        max_steps=50,
        learning_rate=2e-4,
        fp16=True,
        logging_steps=1,
        output_dir="outputs",
        optim="paged_adamw_8bit"
    ),
    peft_config=lora_config,
    formatting_func=formatting_func,
    max_seq_length=1024,
    compute_metrics=compute_metrices
)
```

```python
# training the model
trainer.train()
```

```python
# Model Evaluation
text = "Description: An application keeps crashing. I've tried restarting my computer and reinstalling the application, but it's still crashing. I'm using a Microsoft Office application on a Windows 11 computer."
device = "cuda:0"
inputs = tokenizer(text, return_tensors="pt").to(device)

outputs = model.generate(**inputs, max_new_tokens=20)
output_response = tokenizer.decode(outputs[0], skip_special_tokens=True)
print(output_response)
```

```python
# Save the weights of new trained fine tuned model
new_model = "new_fine_tuned_gemma_model"
trainer.model.save_pretrained(new_model)
tokenizer.save_pretrained(new_model)
model.config.use_cache=True
```
